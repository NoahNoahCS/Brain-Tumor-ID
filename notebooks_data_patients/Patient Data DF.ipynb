# Data Frame 3

```python
# Imort the third data frame

glioma_df3 = pd.read_csv('data/patient_data/difg_msk_2023_clinical_data.tsv', sep='\t') 
list(enumerate(glioma_df3.columns))
```

```python
# Extract the desired features from the third data frame

df_3 = glioma_df3.iloc[: , [6, 7, 37, 25, 33, 5, 28]].copy()
df_3.head()
```

```python
# Adjust the feature names to match primary df

df_3.rename(columns={'WHO Grade':'Tumor Grade','Primary Tumor Site':'Tumor Location','Age at Which Sequencing was Reported (Years)':'Age'}, inplace=True)
df_3.head()
```

```python
# Count null values for all features in the data frame
df_3.isnull().sum()
```

### 'Tumor Grade' feature for df_3

```python
# Determine the unique values of the Tumor Grade feature
df_3['Tumor Grade'].unique()
```

```python
# Edit the values of the Tumor Grade feature to match the values in the primary df
df_3['Tumor Grade'] = df_3['Tumor Grade'].replace('G2', 2)
df_3['Tumor Grade'] = df_3['Tumor Grade'].replace('G3', 3)
df_3['Tumor Grade'] = df_3['Tumor Grade'].replace('G4', 4)

# Check to see if the edits were successful
df_3['Tumor Grade'].unique()
```

```python
# Fill the null values of the Tumor Grade feature 
df_3['Tumor Grade'].fillna(1, inplace=True)

# Change the data type of the Tumor Grade feature to integer
df_3['Tumor Grade'] = df_3['Tumor Grade'].astype(int)

# Check to see if the changes were successful
df_3['Tumor Grade'].unique()
```

```python
# Change data type of Tumor Grade feature to integer

df_3['Tumor Grade'] = df_3['Tumor Grade'].astype(int)
```
